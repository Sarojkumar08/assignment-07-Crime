```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# Import Dataset
crime=pd.read_csv('C:/Users/Saroj Kumar Mohanty/OneDrive/Desktop/exclR assignments/Assignment/7 Clustering/crime_data.csv')
crime
```

```python
crime2=crime.iloc[:,1:]
```

```python
crime2
```

```python
#Normalize the Dataset
crime_norm=pd.DataFrame(normalize(crime2),columns=crime2.columns)
crime_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7)) 
dendrogram=sch.dendrogram(sch.linkage(crime_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(crime_norm),columns=['clusterid'])
y['clusterid'].value_counts()
```

```python
# Adding clusters to dataset
crime2['clustersid']=hclusters.labels_
crime2
```

```python
crime2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
#plot Clusters
plt.figure(figsize=(10,7))
plt.scatter(crime2['clustersid'],crime2['Murder'],c=hclusters.labels_)
```
